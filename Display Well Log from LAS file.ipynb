```python
# extract general header information from the LAS file

well.header
```

```python
# plot the availabe log

well.plot()
```

```python
# displays selected well log, for example GR

tracks=['MD','GR_EDTC']
well.plot(tracks=tracks,lw=0.7)  # lw = line width
```

```python
#display information about the GR log (curve)

well.data['GR_EDTC']
```

```python
# display the start and end value of GR log

well.data['GR_EDTC'].start,well.data['GR_EDTC'].stop
```

```python
# define value at certain depth

well.data['GR_EDTC'].read_at(1500)
```

```python
# print the summary statistic of GR log

well.data['GR_EDTC'].describe()
```

```python
# limit the depth of displayed GR log

well.plot(tracks=[['MD','GR_EDTC']],extents=(1600,2000))
```

```python
# despike GR log

well.data['GR_DESPIKED']=well.data['GR_EDTC'].despike[window_length=70,z=2]
well.plot(tracks=[['GR_EDTC','GR_DESPIKED']])
```

```python
# create color filled version

well.data['GR_EDTC'].plot_2d(cmap='gist_rainbow')
```

```python
# generate subset of GR curve, say between 1400 to 2000 MD

GR_SELECTED=well.data['GR_EDTC'].to_basis(start=1400,stop=2000)
GR_SELECTED.plot()
```

```python
# color the curve defined by the x limit

(GR_SELECTED).plot_2d(cmap='afmhot_r',curve=True,lw=0.6,edgecolor="k")
plt.xlim(0,200)
```
